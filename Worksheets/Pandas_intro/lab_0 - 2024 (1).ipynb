Why does x change if y is the array we're changing? Is there any way to prevent x from changing as well?

***Answer Here***

So, when we assign y = x, we are not creating a new array, we are just creating a new reference to the same array. Therefore, when we change y, x will also change. To prevent x from changing, we can use the copy() method to create a new array with the same values as x.

Below we declare x to be a numpy array that contains four values. Notice how when we print the expression `x>2`, it outputs an array of booleans. 

````python
x = np.array([3, 5, -2, -3])

print( x>2 )
````

What do these boolean values represent? How could we use the expression `x>2` to change all the values that are greater than 2 in x?

***Answer Here***

The boolean values represent the result of the comparison between x and 2. As x is an array, the comparison is done element-wise, and saved as an array of booleans.

To change all the values such that `x > 2`, we can simply use the boolean array to index the array x, and change the values that satisfy the condition (that are True). So for example if we want to change all values such that `x > 2` to `n`, we can do `x[x > 2] = n`.

### Part 3

Define a function called `onlyPositive` that accepts a 1D array as input. 
This function will create a copy of the input array using the `copy()` method from Numpy, use Numpy's fancy indexing to set all negative values in array to 0, and finally return the modified copy.

````python
# Define your function
def onlyPositive(array):
    # create a new array as a copy of previous array
    x = np.copy(array)
    
    # set all negative values in new array to 0
    x[x < 0] = 0

    # return the new array
    return x
````

## Exercise 4

Plot the function $f(x)=x^2 - 3$ on the interval $[-1, 10]$.
Add the label for x and y using Matplotlib's `xlabel()` and `ylabel()`.
Add the legend using `legend()`, and add a title using `title()`.

````python
# Define your f function below
def f(x):
    return x**2 - 3

# Use the linspace function to define values of x below
x1 = np.linspace(-1, 10, 100)

# Plot the function using pyplot below
plt.plot(x1 , f(x1), label='f(x)')

# Add a title
plt.title(r"$f(x) = x^2 - 3$ Graph")

# Add a x-label
plt.xlabel(r"$\forall x \in [-1, 10]$")

# Add a y-label
plt.ylabel(r"$x^2-3$")

# Adding Lengend
plt.legend()
````

## Exercise 5

### Part 1
Plot the functions $f(x)=\frac{x}{x^2 + 2x + 1}$ and $g(x)=\frac{x^2}{2x^2 + 2x - 1}$ on the interval $[-5, 5]$. Be sure to include a **title**, **labels**, and a **legend**. Make sure that $f(x)$ and $g(x)$ are on the same graph.

````python
# Define your f function below
def f(x):
    return x / (x**2 + 2*x + 1)

# Define your g function below
def g(x):
    return x**2 / (2*(x**2) + 2*x + 1)

# Use the linspace function to define values of x below
x2 = np.linspace(-5, 5, 100)

# # Plot the f function using pyplot below
plt.plot(x2 , f(x2))

# # Plot the g function using pyplot below
plt.plot(x2 , g(x2))

# # Add a title
plt.title(r"Graphing $f(x) = \frac{x}{x^2 + 2x + 1}$ and $g(x) = \frac{x^2}{2x^2 + 2x + 1}$")

# # Add a x-label
plt.xlabel(r"$\forall x \in [-5, 5]$")

# # Add a y-label
plt.ylabel(r"f(x) and g(x)")

# # Add a legend
plt.legend(['f(x)' , 'g(x)'])
````

### Part 2
Notice how legend takes in an array as a parameter. Why do you think this is? How would you specify multiple legend names at once?

***Answer Here***

The legend method takes an array as a parameter because it can take multiple labels. To specify multiple legend names at once, we can pass an array of strings as the parameter. Each string in the array will be used as a label for the corresponding plot.
