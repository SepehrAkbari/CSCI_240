## Exercise 1

Compute and plot the quadric approximation of the function $\frac{1}{3x-x^2}$ centered at $1$. Use it to approximate values of $x$ in $[1,3]$

````python
x = symbols('x')
f = 1 / (3*x - x**2)

f_1 = f.subs(x, 1)
f_prime = diff(f, x)
f_prime_1 = f_prime.subs(x, 1)
f_double_prime = diff(f_prime, x)
f_double_prime_1 = f_double_prime.subs(x, 1)

def taylor_approx(x_val):
    return f_1 + f_prime_1 * (x_val - 1) + (f_double_prime_1 / 2) * (x_val - 1)**2

x_vals = np.linspace(1, 3, 100) # 100 is points
f_vals = [1 / (3*n - n**2) for n in x_vals]
approx_vals = [taylor_approx(n) for n in x_vals]

plt.plot(x_vals, f_vals, label='f(x)')
plt.plot(x_vals, approx_vals, label='Apprx')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Quadratic Approximation of $f(x) = \\frac{1}{3x - x^2}$ centered at x=1')
plt.legend()
plt.grid(True)
plt.show()
````
